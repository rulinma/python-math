### 东南西北

```python
import matplotlib.pyplot as plt  

# mat可以认为是math的缩写，plot是画图的意思，lib是库，py是python的缩写

# 创建一个新的figure和axes  
fig, ax = plt.subplots()  
  
# 设置坐标轴的范围（可选）  
ax.set_xlim(-10, 10)  
ax.set_ylim(-10, 10)  
  
# 添加坐标轴标签  
ax.set_xlabel('X')  
ax.set_ylabel('Y')  
  
# 添加标题  
ax.set_title('X-Y Axis')  
  
# 显示网格（可选）  
ax.grid(True)  
  
# 显示图形  
plt.show()
```

### 角

```python
import numpy as np  
import matplotlib.pyplot as plt  
  
# 设置角度，注意这里我们使用弧度制，因为matplotlib内部使用弧度  
angles = np.radians([30, 60, 90, 120, 150, 270])  # 将角度转换为弧度  
  
# 创建一个新的figure和一个极坐标轴  
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})  
  
# 绘制角度对应的点，这里我们简单地将半径设置为1（即圆上）  
# 注意：由于270度（或-90度）与y轴的负方向对齐，所以在圆上它应该与(π, 1)相对应  
# 但由于我们使用了numpy.radians，它已经为我们处理了角度到弧度的转换  
ax.plot(angles, np.ones_like(angles), 'o-')  # 'o-' 表示用圆圈和线连接  
  
# 为每个点添加标签  
for angle, label in zip(angles, ['30°', '60°', '90°', '120°', '150°', '270°']):  
    ax.text(angle, 1.05, label, ha='center', va='center', transform=ax.transData)  
  
# 设置极坐标的theta（角度）标签  
ax.set_thetagrids(np.radians(np.arange(0, 360, 30)), labels=np.arange(0, 360, 30))  
  
# 显示图形  
plt.show()
```

### 三角形与四边形

三角形

```python
import matplotlib.pyplot as plt  
  
# 三角形的顶点坐标  
triangle_points = [(0, 0), (1, 0), (0.5, 0.866)]  
  
# 创建一个新的图形和坐标轴来绘制三角形  
fig_triangle, ax_triangle = plt.subplots()  
ax_triangle.plot(*zip(*triangle_points), 'b-')  # 绘制三角形的边  
ax_triangle.fill(*zip(*triangle_points), 'b', alpha=0.5)  # 填充三角形内部  
ax_triangle.set_title('Triangle')  # 设置标题  
ax_triangle.set_aspect('equal')  # 保持纵横比一致，以便三角形看起来不会变形  
  
# 显示图形  
plt.show()
```

四边形

```python
import matplotlib.pyplot as plt  
  
# 四边形（矩形）的顶点坐标  
quadrilateral_points = [(1, 1), (2, 1), (2, 2), (1, 2)]  
  
# 创建一个新的图形和坐标轴来绘制四边形  
fig_quadrilateral, ax_quadrilateral = plt.subplots()  
ax_quadrilateral.plot(*zip(*quadrilateral_points), 'r-')  # 绘制四边形的边  
ax_quadrilateral.fill(*zip(*quadrilateral_points), 'r', alpha=0.5)  # 填充四边形内部  
ax_quadrilateral.set_title('Quadrilateral')  # 设置标题  
ax_quadrilateral.set_aspect('equal')  # 保持纵横比一致，以便四边形看起来是矩形  
  
# 显示图形  
plt.show()
```

### 流程图

```python
def welcome_by_time(hour):
    if hour < 12:  # 12点之前认为是早上，之后认为是下午
        print("早上好:" , hour)
    else:
        print("下午好:" , hour)


welcome_by_time(8)
welcome_by_time(20)
```

```python
def welcome_by_time(hour):
    if hour < 12:  # 12点之前认为是早上，之后认为是下午
        print("早上好:" , hour)
    elif hour == 12: #elif 可以认为是else if的缩写
        print("中午好:" , hour)
    else:
        print("下午好:" , hour)


welcome_by_time(8)
welcome_by_time(12)
welcome_by_time(20)
```
